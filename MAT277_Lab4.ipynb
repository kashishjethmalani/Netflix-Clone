```python
import numpy as np#numpy arrays are simple than normal python arrays, or it reduces the time complexity
from numpy.random import randint
import matplotlib.pyplot as plt
```

```python
n=1000 #Number of times to run the experiment
die1=randint(low=1, high=7,size=n) #Get list of n random integers from 1 to 6
die1    #Check the generated list
```

```python
np.shape(die1)
```

```python
die2=randint(low=1,high=7,size=n)
```

```python
dice_sum=die1+die2
dice_sum
```

```python
np.shape(dice_sum)
```

```python
n_A=len(dice_sum[dice_sum==5]) #Get number of times the sum was 5
n_A                            #Display the number
```

```python
p_A=n_A/n #Calculate the probability that the sum was 5
p_A       #Display the value of Probability
```

Part2: With functions, for multiple n

```python
def simulate(n,req_sum):
  die1=randint(low=1,high=7,size=n)
  die2=randint(low=1,high=7,size=n)
  dice_sum=die1+die2
  n_A=len(dice_sum[dice_sum==req_sum])
  p_A=n_A/n
  return p_A
```

```python
simulate(1000,5)
```

Now, stimulating for different values of n

```python
p_A_list=[simulate(i,5) for i in range(10,100001,100)]
fig=plt.figure(figsize=(30,10))
plt.plot(range(10,100001,100), [1/9 for i in range(10,100001,100)], color='red', linestyle='dashed')
plt.plot(range(10,100001,100), p_A_list)
plt.xlabel('Number of trials')
plt.ylabel('Probability of event A')
plt.show()
```
